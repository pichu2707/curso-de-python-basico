## Creación de subconjuntos de datos

```python
account_length = data["Account Length"]
account_length.head()
```

```python
# Cuando es una única columna lo que tenemos es una serie
type(account_length)
```

 Una **serie** en pandas es un dato unidimensional, similar a un arraya en numPy o una columna en una tabla, la diferencia de una lista simplle es que tiene una etiqueta sociada al cada elemento Id.

```python
desired_columns = ["Account Length", "Phone", "Eve Calls", "Day Calls"]
subset = data[desired_columns]
subset.head()
```

```python
# Estos son los datos con los que vamos a trabajar ahora
type(subset)
```

```python
all_columns_list = data.columns.values.tolist()
all_columns_list
```

```python
# Podemos crear listas con los bloques que queremos o no trabajar que es lo que vamos a hacer en este caso
# Al tener más de una columna ya nos encontramos con que es un DataFrame

sublist = [x for x in all_columns_list if x not in desired_columns]
sublist
```

```python
# Mostrar la lista del DataFrame a partir del un número incluyendo el primero pero uno menos del último
data[10:30]
```

```python
# Filas con Day Mins que sean menor a 300
data1 = data[data["Day Mins"]<=300]
data1
```

```python
# Buscar filas donde State sea igual a Nueva York (NY)
data2 = data[data["State"]=="NY"]
data2
```

```python
# Creamos un filtro donde buscamos las filas en el que State sea igual a NY y Day Calls menor a 300
data3 = data[(data["State"]=="NY") & (data["Day Mins"]>300)]
data3
```

```python
# Creamos un filtro donde buscamos las filas en el que State sea igual a NY o Day Calls menor a 300
data4 = data[(data["State"]=="NY") | (data["Day Mins"]>300)]
data4
```

```python
# Sacamos las filas en las que Day Calls sean menor que Night Calls

data5 = data[data["Day Calls"]<data["Night Calls"]]
data5
```
